```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
#@author: Cody Jones
#Example orbital code 
C = -105 # a constant which is a combo of the grav const, masses, etc.
timeLimit = 99

# format [x velocity, x position, y velocity, y position]
loc = [2, 10, 2, 1] # initial values

# function used by odeint which returns various derivatives
def R(r, t):
    xaccel = np.cos(np.arctan2(r[3], r[1])) * (C / (r[1]**2 + r[3]**2)**.5); 
    xvel = r[0];
    yaccel = np.sin(np.arctan2(r[3], r[1])) * C / (r[1]**2 + r[3]**2)**.5;
    yvel = r[2];
    
    return np.array([xaccel, xvel, yaccel, yvel])

loc = odeint(R, loc, np.linspace(0, timeLimit, timeLimit + 1));
    
x = loc[:,1]; # extracts the x values from loc
y = loc[:,3]; # extracts the y values from loc

plt.plot(x, y); #plot trajectory
plt.scatter(0, 0, s=400, c = 'r') #center of galaxy
plt.annotate("t = 0", xy = (loc[1, 1], loc[1,3]), xytext=(loc[1, 1] + 2, loc[1,3]- 15), \
            arrowprops=dict(facecolor='black', shrink=0.05, width=.5)); #label start
plt.grid();
plt.show();
```

```python
def Period(a,M):
    """Returns period of orbiting object in Million Years.
    
    Utilizes Kepler's 3rd Law combined with Newton's Law of Universal Gravitation. 
    This equation was taken from ASTR302 Stars and Galaxies"""
    
    d = a*(3.0856776*10**19) #kpc -> meters
    G = 6.673848*10**-11 #N*m^2/kg^2
    M = M*(1.98855*10**30) #Msol -> kg
    #a = a*1000 #kpc -> pc
    #G = 4.302*10**(-3) #pc*(km/s)^2/Msol
    Psec = np.sqrt((4*(np.pi**2)*(d**3))/(G*M)) #in seconds
    P = Psec/(3600*24*365)*10**-6 #Converts seconds to years, 10**-6 converts years to Million Years
    return float(format(P,'.3f'))
```

```python
class Orbital(object): 
    """Orbital Rings for Parts 1-3
    
    This class generates all the relevent data for each of the five rings in the galaxy we are looking at."""
    def __init__(self, name, orbital, position, objects):
        self.name = name
        self.orbital = orbital
        self.position = position #Position is based on Toomre and Toomre regarding each orbital being a certain distance from the eventual two galaxy elliptical pericenter distance
        self.objects = objects #We are told the number of objects in the beginning of Toomre and Toomre II
        
    def description(self):
        r = self.position*R_min
        print "%s is the ring located at %f kpc from the center of the galaxy, with %i objects, ie. Stars, in its ring." %(self.name,r,self.objects)
    
    def picture(self): #Generates an image of the appropriate ring with the correct number of stars and distance from center (which represents the center of the galaxy)
        n = 360/self.objects #Numbers of objects/stars/etc that are defined to be in each ring, based on Toomre and Toomre
        t = np.arange(0.,2*np.pi+0.01,0.01) 
        tstar = np.arange(0.,2*np.pi+0.1,n*np.pi/180) #Used to generate initial positions of the stars
        
        r = self.position*R_min #Each orbital is a certain distance away - this generates that distance
        x = r*np.cos(t)
        y = r*np.sin(t)
        
        xstar = r*np.cos(tstar) #Puts the stars into rings
        ystar = r*np.sin(tstar) #Puts the stars into rings
        
        
        P = Period(r,LargeMass) #Utilizes the above period function
        initVel = (2*np.pi*r)/P #Dist/Time
        vx = initVel*np.cos(tstar) #Initial velocity in x
        vy = initVel*np.sin(tstar) #Intital velcity in y
    
        #plt.plot(x,y,'k-',label = "Orbital Ring") #If we want to draw a circle in addition to the star markers
        plt.plot(xstar,ystar,'bo',markersize = 7,label = 'Stars')
        plt.plot(0,0,'ro')
        plt.xlim(-25,25) #The limits are the R0 distance given in Toomre and Toomre, for the purposes of visualization.
        plt.ylim(-25,25)
        
        X,Y = xstar,ystar
        U,V = vx,vy
        plt.barbs(X,Y,-U,-V,flagcolor='g',barb_increments=dict(half=0.05, full=0.1, flag=0.2),flip_barb=True,length=7,pivot='middle') #This plots a vector field showing the initial rotation of the orbitals.
        
        
    def Stars(self): #Generates an array containing the star positions, and associated velocities, w.r.t. time. 
        r = self.position*R_min
        P = Period(r,LargeMass) #Utilizes the above period function
        initVel = (2*np.pi*r)/P #Dist/Time
        
        n = 360/self.objects #Used to generate stars again
        tstar = np.arange(0.,2*np.pi+0.1,n*np.pi/180+0.1)
        xstar = r*np.cos(tstar)
        ystar = r*np.sin(tstar)
   
        vx = initVel*np.cos(tstar) #Initial velocity in x
        vy = initVel*np.sin(tstar) #Intital velcity in y
        
        return np.array([xstar,ystar,vx,vy]) #The goal is the pass this array into the deriv function, much like we did with the damped projectile motion
```

```python
def GalaxyMotion(b,a):
    """This function plots the overall motion of the center of the two galaxies.
    Based on Toomre and Toomre, we assume the galaxies move in exact parabolic trajectories.
    
    The frequency of the R distances is based on Toomre and Toomre, where they show pictures from t=-1 to t=8 in integer steps.
    Each step corresponds to 10**8 years. t = 0 is the point where both galaxies cross the x-axis 
    
    b is the separation of the galaxies at t=0. b must be a positive number.
    
    a is either 1 or -1, depending on which case of motion we want.
        1: Retrograde (Case 1)
        -1: Direct (Case 2,3,4)
    
    Returns an array with the values of R in its X and Y components."""
    
    plt.figure(figsize=(10,10),dpi=80) #Center axes and make larger
    ax = plt.gca() 
    ax.spines['right'].set_color('none') #From Graphics Tour
    ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.spines['bottom'].set_position(('data',0))
    ax.yaxis.set_ticks_position('left')
    ax.spines['left'].set_position(('data',0))
    
    
    y1 = np.arange(-1.,8.01,0.01) #Used to generate the parabolic arcs, from t=-1 to t=8
    y2 = np.arange(-1,9,1) #Used to generate the integer steps, each of which represents 10**8 years
    
    x = lambda y: y**2-b #Parabolic curve, where b is the separation at t=0
    
    Rx = [] #Lists for x&y components of position
    Ry = []
    vRx = [] #Lists for x&y components of velocity
    vRy = []
    R = []
    
    plt.plot(x(y1),a*y1,'r-',label="Main galaxy") #This the galaxy we are interested in.
    plt.plot(-x(y1),-a*y1,'b-',label="Companion galaxy") #This is the companion galaxy with no disk.
    plt.legend(loc='best')
    
    for i in y2:
        #plt.plot(x(y1),a*y1,'r-',label="Main galaxy") #This the galaxy we are interested in.
        #plt.plot(-x(y1),-a*y1,'b-',label="Companion galaxy") #This is the companion galaxy with no disk.
        plt.plot([x(i),-x(i)],[a*i,-a*i],'g--') #Draws a dashed line from the x&y coordinates of the Main Galaxy to the x&y coords of the Companion Galaxy at each step
        plt.plot(x(i),a*i,'ko') #These two marker plots mimic the markers in Toomre and Toomre, showing the galaxies' positions at different steps. 
        plt.plot(-x(i),-a*i,'ko')
        
        #print i,x(i)
        #print -i,-x(-i)
        Rx.append(np.abs(x(i)+x(i))) #Appends the values of Rx, Ry, vRx, and vRy to appropriate lists
        Ry.append(np.abs(2*i))
        vRx.append(np.abs(x(i+1)-x(i)))
        vRy.append(1)
        
        
    return np.array([Rx,Ry,vRx,vRy]) #This array is the p input array of the below deriv function.
```

```python
def GalaxyTrajectory(b):
    """Simple function that returns JUST the array part of the GalaxyMotion function"""
    y1 = np.arange(-1.,8.01,0.01) #Used to generate the parabolic arcs, from t=-1 to t=8
    y2 = np.arange(-1,9,1) #Used to generate the integer steps
    
    x = lambda y: y**2-b #Parabolic curve, where b is the separation at t=0
    
    Rx = [] #Lists for x&y components of position
    Ry = []
    vRx = [] #Lists for x&y components of velocity
    vRy = []
    for i in y2:
        #print i,x(i)
        #print -i,-x(-i)
        Rx.append(np.abs(x(i)+x(i))) #Appends the values of Rx, Ry, vRx, and vRy to appropriate lists
        Ry.append(np.abs(2*i))
        vRx.append(np.abs(x(i+1)-x(i)))
        vRy.append(1)
    return np.array([Rx,Ry,vRx,vRy]) 
```

1. γ is the Gravitational constant.
2. M is the central mass of the main galaxy and S is the central mass of the disrupting galaxy
3. r is the radius vector from mass M to massless point particle m, representing a single (massless) star in the outer disk of the main galaxy.
4. R is the radius vector from M to S
5. ρ=R−r

```python
def deriv(q,t): 
    gamma = 4.302*10**(-3) #pc*(km/s)^2/Msol
    M = LargeMass
    S = LargeMass
    
    xx = q[0] #These 4 contain details about the orbital rings
    yy = q[1]
    vrx = q[2]
    vry = q[3]
    
    Rx = q[4] #These 4 deal with the overall galaxy motion
    Ry = q[5]
    vRx = q[6]
    vRy = q[7]
    
    r = np.sqrt(xx**2+yy**2)
    R = np.sqrt(Rx**2+Ry**2)
    rho = R-r
    rhox = Rx-xx
    rhoy = Ry-yy
    
    dvrxdt = -gamma*((M/r**3)*xx-(S/rho**3)*rhox+(S/R**3)*Rx)
    dvRxdt = -gamma*((M+S)/R**3)*Rx
    dvrydt = -gamma*((M/r**3)*yy-(S/rho**3)*rhoy+(S/R**3)*Ry)
    dvRydt = -gamma*((M+S)/R**3)*Ry
    
    return (vrx,vry,dvrxdt,dvrydt,vRx,vRy,dvRxdt,dvRydt,)
```

```python
t = np.arange(0.,10.,1)
q = np.array([4.05,2.9,0.24,0.17,48.,2.,1.,1.])
Result = odeint(deriv,q,t)
print Result

```
