This is a simplified simulation of two galaxies merging - simplified in that it only one galaxy has any stars, the motion takes place in a plane, and the 120 stars sampled are massless and thus we ignore a whole slew of complicated relationships involving the gravitational attraction between stars. However, there is enough left to generate an interesting model that allows us to see what happens in a very specific, theoretical, situation.

My personal project was to create a smulation detailing the Direct Passage, Equal Mass Galaxies example from the <a href="http://articles.adsabs.harvard.edu/cgi-bin/nph-iarticle_query?1972ApJ...178..623T&amp;data_type=PDF_HIGH&amp;whole_paper=YES&amp;type=PRINTER&amp;filetype=.pdf"> 1972 Toomre and Toomre paper</a> which involved a circular galaxy,rotating counter-clockwise and passing the companion galaxy in a counter-clockwise manner as well.

<img src="files/DirectPassage.png" />

From Toomre and Toomre, we are given a set of initial conditions that are used in the four examples they present. 
1. The Main Galaxy has 120 stars arranged in 5 rings; they are separated in 12, 18, 24, 30, and 36 stars from the innermost to the outermost ring.
2. Each ring is a specific distance away from the pericenter of the galactic orbit, Rmin, which is set to be 25 kpc.
    * From inner to outer ring, these distances are 0.2, 0.3, 0.4, 0.5, and 0.6 of Rmin, such that the outermost ring orbits at 15 kpc from the center of the Main Galaxy.
3. The mass of the Main Galaxy (and the Companion in this example) is set to be `1e11` Solar Masses.
4. Each integer time step is representative of `1e8` years.
5. The passages are *exactly* parabolic, which I have assumed to mean `y = ax**2` without any widening or narrowing (`a = 1`)

In addition to the Toomre and Toomre paper, we were given a set of coupled differential equations from the German <a href="http://articles.adsabs.harvard.edu/cgi-bin/nph-iarticle_query?1963ZA.....58...12P&amp;data_type=PDF_HIGH&amp;whole_paper=YES&amp;type=PRINTER&amp;filetype=.pdf">Pfleiderer 1963 paper</a> that handle the equations of motion for these galaxies.
```python
"""INITIAL CONDITIONS AND CONSTANTS"""

LargeMass = 1e11 #Solar Masses
#LargeMass = 2e41 #kg
SmallMass = LargeMass
t_increment = 1e8 #Years
R_min = 25 #kpc
#R_min = 7.75e20 m
Orbital5_Dist = 0.6*R_min #15 kpc

G = 4.302e-8 #pc*(km/s)^2/Msol
```

With the inital data out of the way, it is time to begin coding these galaxies. I began by setting up a class of objects to represent the orbital rings and all their stars. This class generated an initial image and the required initial data per star.

```python
class Orbital(object): 
    """Orbital Rings
    
    This class generates all the relevent data for each of the five rings in the galaxy we are looking at.
    The class can also be used to generate a detailed in depth image of the orbital's initial conditions."""
    def __init__(self, name, orbital, position, objects):
        self.name = name
        self.orbital = orbital
        self.position = position #Position is based on Toomre and Toomre regarding each orbital being a certain distance from the eventual two galaxy elliptical pericenter distance
        self.objects = objects #We are told the number of objects in the beginning of Toomre and Toomre II
        
    def description(self):
        r = self.position*R_min
        print "%s is the ring located at %.3f kpc from the center of the galaxy, with %i objects, ie. Stars, in its ring." %(self.name,r,self.objects)
    
    def picture(self): #Generates an image of the appropriate ring with the correct number of stars and distance from center (which represents the center of the galaxy)
        n = 360/self.objects #Numbers of objects/stars/etc that are defined to be in each ring, based on Toomre and Toomre
        t = np.arange(0.,2*np.pi+0.01,0.01) 
        tstar = np.arange(0.,2*np.pi+0.1,n*np.pi/180) #Used to generate initial positions of the stars
        
        r = self.position*R_min #Each orbital is a certain distance away - this generates that distance
        
        #x = r*np.cos(t)
        #y = r*np.sin(t)
        #plt.plot(x,y,'k-',label = "Orbital Ring") #If we want to draw a circle in addition to the star markers
        
        xstar = r*np.cos(tstar) #Puts the stars into rings, in kpc
        ystar = r*np.sin(tstar) #Puts the stars into rings
        
        initVelGMR = -np.sqrt(G*LargeMass/r) #Orbital Velocity = sqrt(GM/R) kpc/YRe8
        vy = -initVelGMR*np.cos(tstar) #Initial velocity in y
        vx = initVelGMR*np.sin(tstar) #Intital velcity in x
    
        plt.plot(xstar,ystar,'bo',markersize = 7,label = 'Stars') #Plots the star locations as blue dots
        plt.plot(0,0,'ro') #Plots the center of the galaxy as a red dot
        
        X,Y = xstar,ystar #USed to create the velocity vectors
        U,V = vx,vy
        plt.barbs(X,Y,U,V,flagcolor='g',barb_increments=dict(half=4, full=9, flag=15),flip_barb=False,length=7, pivot = 'middle') #This plots a vector field showing the initial rotation of the orbitals.
        
        
    def Stars(self): #Generates an array containing the star positions, and associated velocities, w.r.t. time. 
        r = self.position*R_min

        initVelGMR = -np.sqrt(G*LargeMass/r) #kpc/YRe8
        n = 360/self.objects #Used to generate stars again
        tstar = np.arange(0.,2*np.pi,n*np.pi/180)
        xstar = r*np.cos(tstar) #Star Positions in kpc
        ystar = r*np.sin(tstar)
   
        vy = -initVelGMR*np.cos(tstar) #Initial velocity in y
        vx = initVelGMR*np.sin(tstar) #Intital velcity in x
        
        return np.array([xstar,ystar,vx,vy]) #This array contains all the initial data for each star in each ring.
```

However, that is just one ring, so to get an idea of the full thing, we can put them all together to generate the Main Galaxy.

```python
def Galaxy(i):
    """Generates a galaxy with i rings based on the Orbital Class.
    
    Displays a graph with all the orbital rings on one graph, generating an image of the simplified circular galaxy."""
    plt.figure(figsize=(10,10),dpi=80) #Makes the graph large and square to properly show our mock circular galaxy
    ax = plt.gca()
    ax.tick_params(direction='out',top=False,right=False) #Cleans up the graph axes
    ax.spines['top'].set_visible(False) 
    ax.spines['right'].set_visible(False)
    
    plt.xlabel('X (kpc)') #The distances are measured in kpc, with the center at the origin. 
    plt.ylabel('Y (kpc)')
    plt.title('Galaxy With %i Rings' %(i))


    for j in range(1,i+1): #For each ring up to the specified ring
        Orbital_j = Orbital("Orbital %s" %(j),j,(0.2+(j-1)*0.1),12+(j-1)*6) #Generates a ring based on the Orbital Class that has the proper characteristics based on ring number j
        Orbital_j.picture() #Adds the orbital image to the graph
        Orbital_j.Stars() #Generates the relevant data for each ring
        
            
Galaxy(5)
```

Using similar code, I created a list to arrange as an array containing all the information for each star in the galaxy.

```python
def InitialConditions(i):
    """Generates a galaxy with i rings based on the Orbital Class.
    
    Returns all initial data for each star in a galaxy of i rings."""
    Ox = [] #Empty lists for each x, y, vx, and vy of each star.
    Oy = []
    Ovx = []
    Ovy = []

    for j in range(1,i+1): 
        Orbital_j = Orbital("Orbital %s" %(j),j,(0.2+(j-1)*0.1),12+(j-1)*6) #Generates a ring based on the Orbital Class that has the proper characteristics based on ring number j
        Orbital_j.Stars() #Generates the relevant data for each ring
        Ox_j = Orbital_j.Stars()[0] #Takes the first subarray (x positions) from the Star orbital function
        Oy_j = Orbital_j.Stars()[1] #y positions
        Ovx_j = Orbital_j.Stars()[2] #vx components
        Ovy_j = Orbital_j.Stars()[3] #vy compnents 
        
        for a in range(12+(j-1)*6): #The range is the number of stars in each orbital
            Ox.append(Ox_j[a]) #Appends each value in the respective array to the respective list, which is easier to work with for future manipulation than arrays
            Oy.append(Oy_j[a])
            Ovx.append(Ovx_j[a])
            Ovy.append(Ovy_j[a])
    return Ox, Oy, Ovx, Ovy #Returns the lists
        
```

```python
def GalaxyMotion(b,a):
    """This function plots the overall motion of the center of the two galaxies.
    Based on Toomre and Toomre, we assume the galaxies move in exact parabolic trajectories.
    
    The frequency of the R distances is based on Toomre and Toomre, where they show pictures from t=-1 to t=8 in integer steps.
    Each step corresponds to 10**8 years. t = 0 is the point where both galaxies cross the x-axis 
    
    b is the separation of the galaxies at t=0. b must be a positive number.
    
    a is either 1 or -1, depending on which case of motion we want.
        1: Retrograde (Case 1)
        -1: Direct (Case 2,3,4)
    
    Returns an array with the values of R in its X and Y components."""
    
    plt.figure(figsize=(10,10),dpi=80) #Center axes and make larger
    ax = plt.gca() 
    ax.spines['right'].set_color('none') #From Graphics Tour
    ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.spines['bottom'].set_position(('data',0))
    ax.yaxis.set_ticks_position('left')
    ax.spines['left'].set_position(('data',0))
    
    
    y1 = np.arange(-1.,8.01,0.01) #Used to generate the parabolic arcs, from t=-1 to t=8
    y2 = np.arange(-1,9,1) #Used to generate the integer steps, each of which represents 10**8 years
    
    x = lambda y: y**2-b #Parabolic curve, where b is the separation at t=0
    
    Rx = [] #Lists for x&y components of position
    Ry = []
    vRx = [] #Lists for x&y components of velocity - kpc/YRe8
    vRy = []
    
    plt.plot(x(y1),a*y1,'r-',label="Main galaxy") #This the galaxy we are interested in.
    plt.plot(-x(y1),-a*y1,'b-',label="Companion galaxy") #This is the companion galaxy with no disk.
    plt.legend(loc='best')
    
    
    for i in y2:
        plt.plot([x(i),-x(i)],[a*i,-a*i],'g--') #Draws a dashed line from the x&y coordinates of the Main Galaxy to the x&y coords of the Companion Galaxy at each step
        plt.plot(x(i),a*i,'ko') #These two marker plots mimic the markers in Toomre and Toomre, showing the galaxies' positions at different steps. 
        plt.plot(-x(i),-a*i,'ko')
        plt.title("Flat Direct Parabolic Passage")
        
        Rx.append(np.abs(x(i)+x(i))) #Appends the values of Rx, Ry, vRx, and vRy to appropriate lists
        Ry.append(np.abs(2*i))
        vRx.append(np.abs(x(i+1)-x(i)))
        vRy.append(1)
        
        
    return np.array([Rx,Ry,vRx,vRy]) #This array is the p input array of the below deriv function.
```

Unfortunately, we aren't interested in pure passage, but rather in what happens to two galaxies under the constraint of the above equations. To do that, we need to use numpy's `odeint` function. The code itself is quite similar to the code used in our damped projectile motion exercise.

```python
def Rderiv(q,t): 
    """
    A function used in conjunction with odeint to solve the R-acceleration equation above. 
    """
    gamma = G
    M = LargeMass
    S = LargeMass
   
    Rx = q[0] #Pass the first value in the array as Rx, etc...
    Ry = q[1]
    vRx = q[2]
    vRy = q[3]

    R = np.sqrt(Rx**2+Ry**2) #Pythagorean Theorem to get R
    
    dvRxdt = -gamma*((M+S)/R**3)*Rx #The equation is split into x and y components, as the bolded R in the equation represents a vector.
    dvRydt = -gamma*((M+S)/R**3)*Ry
    
    return (vRx,vRy,dvRxdt,dvRydt) #Returns the outputted values, which when run through odeint will be [Rx,Ry,vRx,vRy]
```

```python
def Rorbit(time=0.1):
    """
    An interactive function that shows the resulting orbit of two galaxies, and compares it to the Toomre passages.
    """
    t = np.arange(0.,time,.1) #Interval of .1 ~ 1e7 yrs
    q = np.array([50.,0.,1.,1.]) #Initial conditions taken from the above static picture of the parabolic curves, starting at t = 0
    
    Result = odeint(Rderiv,q,t) #odeint with passed Rderiv and conditions

    Rx = Result[:,0] #Pulling out the x and y components from the derivative 
    Ry = Result[:,1]
    
    
    plt.figure(figsize=(10,10),dpi=80) #Setting up the graph
    ax = plt.gca() 
    ax.spines['right'].set_color('none') #From Graphics Tour
    ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.spines['bottom'].set_position(('data',0))
    ax.yaxis.set_ticks_position('left')
    ax.spines['left'].set_position(('data',0))
    

    plt.plot(25-Rx,-Ry,'yo',label = "Main Galaxy Orbit") #These two curves graph the orbit of the two galaxies according to the Pfleiderer
    plt.plot(-25+Rx,Ry,'go',label="Companion Galaxy Orbit") #equation, with the correct counter-clockwise rotation, with the correct offset from the center of mass
    
    
    plt.plot(-25,0,'ro') #Initial Main position
    plt.plot(25,0,'bo') #Initial Companion position
    
    
    x = lambda y: y**2-25 #Parabolic function
    
    plt.plot(x(t),-t,'y-',label = "Main Galaxy Passage") #These two lines are the same as the above static picture, showing the projected Toomre Passages
    plt.plot(-x(t),t,'g-',label = "Companion Galaxy Passage") #for the two galaxies
    
    plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.06),
          fancybox=True, shadow=True, ncol=5)
    plt.title('Flat Direct Orbit - Center of Mass Frame')
    plt.show()
    
v = interact(Rorbit,time=(0.,10.1,.1)) #Interval of .1 ~ 1e7 yrs
```

Now that it is seen what happens to the galaxies on the large scale, what happens to the stars in the Main Galaxy? 

For that, we need to code the other equation in. First, it is beneficial to check that the stars actually orbit in a circle without the Companion Galaxy affecting them.

```python
def rderiv(q,t): 
    """
    This derivative function returns the data required to model the Main Galaxy when there is no Companion Galaxy interacting with it. 
    
    Should result in circular orbits.
    """
    gamma = G
    M = LargeMass
    S = 0
    
    xx = q[0] #These 4 contain details about the orbital rings
    yy = q[1] #Pass the first 4 elements of the array into these
    vrx = q[2]
    vry = q[3]
    
    Rx = q[4] #These 4 deal with the overall galaxy motion
    Ry = q[5] #Pass the last 4 elements into these
    vRx = q[6]
    vRy = q[7]
    
    r = np.sqrt(xx**2+yy**2) #Pythagorean Theorem
    R = np.sqrt(Rx**2+Ry**2)
    
    rho = R-r #Rho is defined above 
    rhox = Rx-xx #Rho x and y components
    rhoy = Ry-yy
    
    dvrxdt = -gamma*(((M/r**3)*xx)-((S/rho**3)*rhox)+((S/R**3)*Rx)) #The functions are broken down into their x and y vector components
    dvrydt = -gamma*(((M/r**3)*yy)-((S/rho**3)*rhoy)+((S/R**3)*Ry)) #r x and y - the last two terms in each are 0, S = 0
    dvRxdt = -gamma*((M+S)/R**3)*Rx #R x and y
    dvRydt = -gamma*((M+S)/R**3)*Ry
    
    return (vrx,vry,dvrxdt,dvrydt,vRx,vRy,dvRxdt,dvRydt) #Returned array corresponds to [rx,ry,vrx,vry,Rx,Ry,vRx,vRy]
```

```python
def rorbit(time = 0.1):
    """
    Function showing the circular orbits of the Main Galaxy without a Companion Galaxy present.
    """
    t = np.arange(0.,time,.1) #Interval of .1 ~ 1e7 yrs
    plt.figure(figsize=(10,10),dpi=80)
    for i in range(0,120): #For each of the 120 arrays in the init array.
        Res = odeint(rderiv,initarr[i],t) #Running odeint for each array
        x = Res[:,0] #Pulls out the x and y components from each run through the odeint
        y = Res[:,1]
        
        plt.plot(x,y,'bo',markersize=1) #Plots each star as a small marker

    plt.plot(0,0,'ro', label = 'Center of Main Galaxy')
    plt.xlim(-20,20)
    plt.ylim(-20,20)
    plt.xlabel('X (kpc)')  
    plt.ylabel('Y (kpc)')
    plt.title("Stars in Orbit")
    plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.06),
          fancybox=True, shadow=True, ncol=5)
    plt.show()
v = interact(rorbit,time=(0.,10.1,.1)) #Interval of .1 ~ 1e7 yrs
```
