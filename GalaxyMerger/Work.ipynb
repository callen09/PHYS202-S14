```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from IPython.html.widgets import interact,interactive
```

```python
def deriv2(q,t): 
    gamma = 4.302e-8 #pc*(km/s)^2/Msol
    M = 1e11
    S = 0.25*1e11
    
    
    Rx = q[0] #These 4 deal with the overall galaxy motion
    Ry = q[1]
    vRx = q[2]
    vRy = q[3]

    R = np.sqrt(Rx**2+Ry**2)
    
    dvRxdt = -gamma*((M+S)/R**3)*Rx
    dvRydt = -gamma*((M+S)/R**3)*Ry
    
    return (vRx,vRy,dvRxdt,dvRydt)
```

```python
def Rorbit(time):
    t2 = np.arange(0.,time,.1)
    q2 = np.array([24.,2.,1.,1.])
    Result = odeint(deriv2,q2,t2)

    x2 = Result[:,0]
    y2 = Result[:,1]

    plt.plot(x2,y2,'r-')
    plt.xlabel("X distance between galaxies (kpc)",fontsize=15)
    plt.ylabel("Y distance between galaxies (kpc)",fontsize=15)
    plt.show()
v = interact(Rorbit,time=(0.,10.0))
```

```python
def deriv(q,t): 
    gamma = 4.302e-8 #pc*(km/s)^2/Msol
    M = 1e11
    S = 0.25*1e11
    
    xx = q[0] #These 4 contain details about the orbital rings
    yy = q[1]
    vrx = q[2]
    vry = q[3]
    
    Rx = q[4] #These 4 deal with the overall galaxy motion
    Ry = q[5]
    vRx = q[6]
    vRy = q[7]
    
    r = np.sqrt(xx**2+yy**2)
    R = np.sqrt(Rx**2+Ry**2)
    
    rho = R-r
    rhox = Rx-xx
    rhoy = Ry-yy
    
    dvrxdt = -gamma*(((M/r**3)*xx)-((S/rho**3)*rhox)+((S/R**3)*Rx))
    dvrydt = -gamma*(((M/r**3)*yy)-((S/rho**3)*rhoy)+((S/R**3)*Ry))
    dvRxdt = -gamma*((M+S)/R**3)*Rx
    dvRydt = -gamma*((M+S)/R**3)*Ry
    
    return (vrx,vry,dvrxdt,dvrydt,vRx,vRy,dvRxdt,dvRydt)
```

```python
t = np.arange(0.,10.,1)
q = np.array([4.05,2.9,0.24,0.17,48.,2.,1.,1.])
Result = odeint(deriv,q,t, full_output=1)
print Result
```
