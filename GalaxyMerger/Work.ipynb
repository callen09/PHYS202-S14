```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from IPython.html.widgets import interact,interactive
```

```python
def deriv2(q,t): 
    gamma = 4.302e-8 #pc*(km/s)^2/Msol
    M = 1e11
    S = 1e11
   
    Rx = q[0] #These 4 deal with the overall galaxy motion
    Ry = q[1]
    vRx = q[2]
    vRy = q[3]

    R = np.sqrt(Rx**2+Ry**2)
    
    dvRxdt = -gamma*((M+S)/R**3)*Rx
    dvRydt = -gamma*((M+S)/R**3)*Ry
    
    return (vRx,vRy,dvRxdt,dvRydt)
```

```python
def Rorbit(time):
    t2 = np.arange(0.,time,.1)
    y1 = np.arange(0.,time,.1)
    q2 = np.array([50.,0.,1.,1.])
    Result = odeint(deriv2,q2,t2)

    x2 = Result[:,0]
    y2 = Result[:,1]
    
    
    plt.figure(figsize=(10,10),dpi=80)
    ax = plt.gca() 
    ax.spines['right'].set_color('none') #From Graphics Tour
    ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.spines['bottom'].set_position(('data',0))
    ax.yaxis.set_ticks_position('left')
    ax.spines['left'].set_position(('data',0))
    

    plt.plot(25-x2,-y2,'yo',label = "Main Galaxy Trajectory")
    plt.plot(-25+x2,y2,'go',label="Companion Galaxy Trajectory")
    
    
    plt.plot(-25,0,'ro') #Initial Main
    plt.plot(25,0,'bo') #Initial Companion
    
    
    x = lambda y: y**2-25
    plt.plot(x(y1),-y1,'y-',label = "Main Galaxy Parabolic")
    plt.plot(-x(y1),y1,'g-',label = "Companion Galaxy Parabolic")
    
    plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.06),
          fancybox=True, shadow=True, ncol=5)
    plt.title('Flat Direct Orbit - Center of Mass Frame')
    plt.show()
    
v = interact(Rorbit,time=(0.,10.1,.1))
```

```python
def deriv(q,t): 
    gamma = 4.302e-8 #pc*(km/s)^2/Msol
    M = 1e11
    S = 0
    
    xx = q[0] #These 4 contain details about the orbital rings
    yy = q[1]
    vrx = q[2]
    vry = q[3]
    
    Rx = q[4] #These 4 deal with the overall galaxy motion
    Ry = q[5]
    vRx = q[6]
    vRy = q[7]
    
    r = np.sqrt(xx**2+yy**2)
    R = np.sqrt(Rx**2+Ry**2)
    
    rho = R-r
    rhox = Rx-xx
    rhoy = Ry-yy
    
    dvrxdt = -gamma*(((M/r**3)*xx)-((S/rho**3)*rhox)+((S/R**3)*Rx))
    dvrydt = -gamma*(((M/r**3)*yy)-((S/rho**3)*rhoy)+((S/R**3)*Ry))
    dvRxdt = -gamma*((M+S)/R**3)*Rx
    dvRydt = -gamma*((M+S)/R**3)*Ry
    
    return (vrx,vry,dvrxdt,dvrydt,vRx,vRy,dvRxdt,dvRydt)
```

```python
LargeMass = 10**11 #Solar Masses
#LargeMass = 2e41 #kg
SmallMass = LargeMass
t_increment = 10**8 #Years
R_min = 25 #kpc, given in Toomre and Toomre
#R_min = 7.75e20 m
Orbital5_Dist = 0.6*R_min
Orbital5_Period = 5.442 #Time Units
#OuterPeriod = 544.2e6 Years

class Orbital(object): 
    """Orbital Rings for Parts 1-3
    
    This class generates all the relevent data for each of the five rings in the galaxy we are looking at."""
    def __init__(self, name, orbital, position, objects):
        self.name = name
        self.orbital = orbital
        self.position = position #Position is based on Toomre and Toomre regarding each orbital being a certain distance from the eventual two galaxy elliptical pericenter distance
        self.objects = objects #We are told the number of objects in the beginning of Toomre and Toomre II
        
    def description(self):
        r = self.position*R_min
        print "%s is the ring located at %f kpc from the center of the galaxy, with %i objects, ie. Stars, in its ring." %(self.name,r,self.objects)
    
    def picture(self): #Generates an image of the appropriate ring with the correct number of stars and distance from center (which represents the center of the galaxy)
        n = 360/self.objects #Numbers of objects/stars/etc that are defined to be in each ring, based on Toomre and Toomre
        t = np.arange(0.,2*np.pi+0.01,0.01) 
        tstar = np.arange(0.,2*np.pi+0.1,n*np.pi/180) #Used to generate initial positions of the stars
        
        r = self.position*R_min #Each orbital is a certain distance away - this generates that distance
        x = r*np.cos(t)
        y = r*np.sin(t)
        
        xstar = r*np.cos(tstar) #Puts the stars into rings
        ystar = r*np.sin(tstar) #Puts the stars into rings
        
        
        #P = Period(r,LargeMass) #Utilizes the above period function
        #initVel = (2*np.pi*r)/P #Dist/Time
        #initVel2 = initVel*1e2
        initVelGMR = -np.sqrt(4.302e-8*1e11/r) #kpc/YRe8
        vy = -initVelGMR*np.cos(tstar) #Initial velocity in y
        vx = initVelGMR*np.sin(tstar) #Intital velcity in x
    
        #plt.plot(x,y,'k-',label = "Orbital Ring") #If we want to draw a circle in addition to the star markers
        plt.plot(xstar,ystar,'bo',markersize = 7,label = 'Stars')
        plt.plot(0,0,'ro')
        
        
        X,Y = xstar,ystar
        U,V = vx,vy
        plt.barbs(X,Y,U,V,flagcolor='g',barb_increments=dict(half=4, full=9, flag=15),flip_barb=False,length=7,pivot='middle') #This plots a vector field showing the initial rotation of the orbitals.
        
        
    def Stars(self): #Generates an array containing the star positions, and associated velocities, w.r.t. time. 
        r = self.position*R_min
        #P = Period(r,LargeMass) #Utilizes the above period function
        #initVel = (2*np.pi*r)/P #Dist/Time - kpc/YRe6
        initVelGMR = -np.sqrt(4.302e-8*1e11/r) #kpc/YRe8
        #initVel2 = initVel*1e2 #To get the velocity of the stars onto the same scale as the galactic orbit velocity scale, need to multiply by 10**2
        
        
        n = 360/self.objects #Used to generate stars again
        tstar = np.arange(0.,2*np.pi,n*np.pi/180)
        xstar = r*np.cos(tstar)
        ystar = r*np.sin(tstar)
   
        vy = -initVelGMR*np.cos(tstar) #Initial velocity in y
        vx = initVelGMR*np.sin(tstar) #Intital velcity in x
        
        return np.array([xstar,ystar,vx,vy]) #The goal is the pass this array into the deriv function, much like we did with the damped projectile motion
```
