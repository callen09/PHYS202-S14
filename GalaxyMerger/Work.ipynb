```python
%pylab inline
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from IPython.html.widgets import interact,interactive
```

```python
def deriv2(q,t): 
    gamma = 4.302e-8 #pc*(km/s)^2/Msol
    M = 1e11
    S = 1e11
   
    Rx = q[0] #These 4 deal with the overall galaxy motion
    Ry = q[1]
    vRx = q[2]
    vRy = q[3]

    R = np.sqrt(Rx**2+Ry**2)
    
    dvRxdt = -gamma*((M+S)/R**3)*Rx
    dvRydt = -gamma*((M+S)/R**3)*Ry
    
    return (vRx,vRy,dvRxdt,dvRydt)
```

```python
def Rorbit(time):
    t2 = np.arange(0.,time,.1)
    y1 = np.arange(0.,time,.1)
    q2 = np.array([50.,0.,1.,1.])
    Result = odeint(deriv2,q2,t2)

    x2 = Result[:,0]
    y2 = Result[:,1]
    
    
    plt.figure(figsize=(10,10),dpi=80)
    ax = plt.gca() 
    ax.spines['right'].set_color('none') #From Graphics Tour
    ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.spines['bottom'].set_position(('data',0))
    ax.yaxis.set_ticks_position('left')
    ax.spines['left'].set_position(('data',0))
    

    plt.plot(25-x2,-y2,'yo',label = "Main Galaxy Trajectory")
    plt.plot(-25+x2,y2,'go',label="Companion Galaxy Trajectory")
    
    
    plt.plot(-25,0,'ro') #Initial Main
    plt.plot(25,0,'bo') #Initial Companion
    
    
    x = lambda y: y**2-25
    plt.plot(x(y1),-y1,'y-',label = "Main Galaxy Parabolic")
    plt.plot(-x(y1),y1,'g-',label = "Companion Galaxy Parabolic")
    
    plt.legend(loc='upper center', bbox_to_anchor=(0.5, -0.06),
          fancybox=True, shadow=True, ncol=5)
    plt.title('Flat Direct Orbit - Center of Mass Frame')
    plt.show()
    
v = interact(Rorbit,time=(0.,10.1,.1))
```

```python
def deriv(q,t): 
    gamma = 4.302e-8 #pc*(km/s)^2/Msol
    M = 1e11
    S = 0
    
    xx = q[0] #These 4 contain details about the orbital rings
    yy = q[1]
    vrx = q[2]
    vry = q[3]
    
    Rx = q[4] #These 4 deal with the overall galaxy motion
    Ry = q[5]
    vRx = q[6]
    vRy = q[7]
    
    r = np.sqrt(xx**2+yy**2)
    R = np.sqrt(Rx**2+Ry**2)
    
    rho = R-r
    rhox = Rx-xx
    rhoy = Ry-yy
    
    dvrxdt = -gamma*(((M/r**3)*xx)-((S/rho**3)*rhox)+((S/R**3)*Rx))
    dvrydt = -gamma*(((M/r**3)*yy)-((S/rho**3)*rhoy)+((S/R**3)*Ry))
    dvRxdt = -gamma*((M+S)/R**3)*Rx
    dvRydt = -gamma*((M+S)/R**3)*Ry
    
    return (vrx,vry,dvrxdt,dvrydt,vRx,vRy,dvRxdt,dvRydt)
```

```python
def GalaxyMotion(time):
    """This function plots the overall motion of the center of the two galaxies.
    Based on Toomre and Toomre, we assume the galaxies move in exact parabolic trajectories.
    
    The frequency of the R distances is based on Toomre and Toomre, where they show pictures from t=-1 to t=8 in integer steps.
    Each step corresponds to 10**8 years. t = 0 is the point where both galaxies cross the x-axis 
    
    b is the separation of the galaxies at t=0. b must be a positive number.
    
    a is either 1 or -1, depending on which case of motion we want.
        1: Retrograde (Case 1)
        -1: Direct (Case 2,3,4)
    
    Returns an array with the values of R in its X and Y components."""
    
    plt.figure(figsize=(10,10),dpi=80) #Center axes and make larger
    ax = plt.gca() 
    ax.spines['right'].set_color('none') #From Graphics Tour
    ax.spines['top'].set_color('none')
    ax.xaxis.set_ticks_position('bottom')
    ax.spines['bottom'].set_position(('data',0))
    ax.yaxis.set_ticks_position('left')
    ax.spines['left'].set_position(('data',0))
    #plt.xlim(-50,50)
    #plt.ylim(-50,50)
    
    y1 = np.arange(-1.,time,0.01) #Used to generate the parabolic arcs, from t=-1 to t=8
    y2 = np.arange(-1,time,1) #Used to generate the integer steps, each of which represents 10**8 years
    
    x = lambda y: y**2-25 #Parabolic curve, where b is the separation at t=0
    
    t = np.arange(0.,2*np.pi+0.01,0.01) 
    tstar = np.arange(0.,2*np.pi+0.1,n*np.pi/180)
    r = .2*25
    xr = r*np.cos(t)
    xstar = r*np.cos(tstar)
    yr = r*np.sin(t)
    ystar = r*np.sin(tstar)
    
    Rx = [] #Lists for x&y components of position
    Ry = []
    vRx = [] #Lists for x&y components of velocity
    vRy = []
    R = []
    
    plt.plot(x(y1),-y1,'r-',label="Main galaxy") #This the galaxy we are interested in.
    plt.plot(-x(y1),y1,'b-',label="Companion galaxy") #This is the companion galaxy with no disk.
    plt.legend(loc='best')
    
    for i in y2:
        xstar = r*np.cos(tstar)+x(i) 
        ystar = r*np.sin(tstar)+-i
        #plt.plot(x(y1),a*y1,'r-',label="Main galaxy") #This the galaxy we are interested in.
        #plt.plot(-x(y1),-a*y1,'b-',label="Companion galaxy") #This is the companion galaxy with no disk.
        plt.plot([x(i),-x(i)],[-i,i],'g--') #Draws a dashed line from the x&y coordinates of the Main Galaxy to the x&y coords of the Companion Galaxy at each step
        plt.plot(x(i),-i,'ko') #These two marker plots mimic the markers in Toomre and Toomre, showing the galaxies' positions at different steps. 
        plt.plot(-x(i),i,'ko')
        #plt.plot(xr,yr,'k-')
        #plt.plot(xstar,ystar,'bo',markersize = 7,label = 'Stars')
        
        #print i,x(i)
        #print -i,-x(-i)
        Rx.append(np.abs(x(i)+x(i))) #Appends the values of Rx, Ry, vRx, and vRy to appropriate lists
        Ry.append(np.abs(2*i))
        vRx.append(np.abs(x(i+1)-x(i)))
        vRy.append(1)
        
        
    return np.array([Rx,Ry,vRx,vRy]) #This array is the p input array of the below deriv function.

v = interact(GalaxyMotion,time=(-1.,10.1))
```
