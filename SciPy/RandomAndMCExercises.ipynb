```python
x = np.arange(0.,1.0001,0.0001)
f = lambda x: 60*(x**3)*(1-x)**2
F = np.cumsum(f(x))/f(x).sum()

xmin = 0
xmax = 1.
N = 10000
xsamples = np.zeros(N,float)

for i in range(N):
    rr = np.random.uniform()
    xsamples[i] = x[F.searchsorted(rr)]

plt.plot(x,f(x),lw=3,label="True Distribution")
plt.plot(x,F,lw=3,label = 'Cumulative Distribution')
plt.hist(xsamples,100,normed=True,label="Sampled distribution")


plt.xlabel(r'$x$', fontsize = 20)
plt.xlim(0.,1.)

plt.ylim(0.,3.)
plt.legend(loc='best')
plt.show()
```

```python
def WoodsSaxon(R,w,a,radius):
    r = np.arange(0.,radius+0.1,0.1)
    rho0 = 1
    rho = lambda r: (rho0*(1+w*r**2/R**2))/(1+np.exp((r-R)/a))
    

    plt.plot(r,rho(r),label='$p(r)$',lw=2)
    plt.legend(loc=0)
    plt.show()
    
    plt.plot(r,rho(r)*r**2,label='$p(r)*r^2$',lw=2)
    plt.legend(loc=0)
    plt.show()
```

```python
from scipy.integrate import trapz,cumtrapz
```

```python
def distribute1D(x,prob,N):
    """takes any distribution which is directly proportional 
    to the number of particles, and returns data that is 
    statistically the same as the input data."""
    CDF = cumtrapz(prob)/np.sum(prob)
    xsamples = np.zeros(N,float)
    for i in range(0,N):
        r = np.random.ranf()
        xsamples[i] = x[CDF.searchsorted(r)]
    return xsamples
```

```python
def WoodsSaxonA(R,w,a,N):
    r = np.arange(0.,N+0.1,0.1)
    rho0 = 1
    rho = lambda r: (rho0*(1+w*r**2/R**2))/(1+np.exp((r-R)/a))
    rhor = rho(r)*r**2
    
    A = distribute1D(r,rhor,N)
    B = distribute1D(r,rhor,N)
    
    ysamples = np.random.ranf(N)
    
    u = np.random.uniform(0,1.)
    v = np.random.uniform(0,1.)
    
    phi = 2*np.pi*u
    theta = np.arccos(2*v-1)
    
    dv = r**2*np.sin(theta)
    
    fig = plt.figure(1,(10,6))
    plt.subplot(3,1,1)
    plt.plot(r,rhor,label='$p(r)*r^2$',lw=2)
    plt.xlim(0.,18.)
    
    plt.subplot(3,1,2)
    plt.scatter(A,ysamples,color = 'r',alpha =0.25)
    plt.scatter(B,ysamples,color = 'g',alpha = 0.25)
    plt.xlim(0.,18.)
    
    plt.subplot(3,1,3)
    plt.plot(r,dv)
    
    
    plt.show()
```
