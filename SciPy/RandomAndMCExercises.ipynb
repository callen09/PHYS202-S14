```python
alpha = 1
beta = 1.
x = np.arange(0.,10000.,0.1)

f = lambda x: alpha*(beta**(-alpha))*(x**(alpha-1))*np.exp(-(x/beta)**alpha)
F = lambda x: np.exp(0)-np.exp(-(x/beta)**alpha)

plt.plot(x,f(x),label='True distribution',lw=2)
plt.plot(x,F(x),label='Cumulative function',lw=2)

plt.xlim(0.,10.)
plt.ylim(0.,1.)

plt.legend(loc=0)
plt.show()
```

```python
alpha = 1.5
beta = 6.
N = 10000.
z = np.random.random_sample(N);
x = np.arange(0.,10000.,0.1)

f = lambda x: alpha*(beta**(-alpha))*(x**(alpha-1))*np.exp(-(x/beta)**alpha)
F = lambda x: np.exp(0)-np.exp(-(x/beta)**alpha)
Finv = lambda z: -np.log(1-z**alpha)*beta**alpha 

    
xsamples = Finv(z);


plt.plot(x,f(x),label='True distribution',lw=2)
plt.plot(x,F(x),label='Cumulative function',lw=2)
plt.hist(xsamples,500,normed=True,label="Sampled distribution")

plt.xlim(0.,10.)
plt.ylim(0.,1.)

plt.legend(loc=0)
plt.show()
```

<p><strong>(a)</strong> Write a program that uses the <em>Rejection Sampling Method</em> to generate a set of 10,000 samples from this distribution and plot the density distribution, the CDF, and a histogram of the sampled distribution on the same graph.</p>

```python
f = lambda x: 60*(x**3)*(1-x)**2
F = lambda x: 10*x**6-24*x**5+15*x**4
Finv = lambda x: 1/(10*x**6-24*x**5+15*x**4)

x = np.arange(0.,1.01,0.01)

xmin = 0
xmax = 1.

#find the maximum value of f within bounds
subx = x[logical_and((x>xmin),(x<xmax))]
ymax = f(subx).max()*1.05
ymin = 0.

N = 10000
samples_x = xmin + (xmax-xmin)*np.random.random_sample(N);
samples_y = ymin + (ymax-ymin)*np.random.random_sample(N);
z = np.random.random_sample(N);

newmask = (samples_y < f(samples_x))
newmask.sum()

xsamples = Finv(z);

plt.plot(x,f(x),lw=3,label="True Distribution")
plt.plot(x,F(x),lw=3,label = 'Cumulative Distribution')
#plt.hist(xsamples,50,normed=True,label="Sampled distribution")
plt.plot(samples_x[newmask[:500]],samples_y[newmask[:500]],'r.',label = 'Density Distribution')

#plt.ylabel(r'$x^{1/2} + \sin(x)$',fontsize=20)
plt.xlabel(r'$x$', fontsize = 20)
plt.xlim(0.,1.)
plt.ylim(0.,3.)
plt.show()
```
