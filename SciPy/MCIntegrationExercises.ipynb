**(a)** Use the sample mean method to compute the potential at point P(2,3) due to the disk of charge for different numbers of samples N=(101,102,103,104,105,106). Report the value and its statistical uncertainty for each sample size. How does the result depend on N?

```python
def sampleMean(N):

    #Set up the function
    Xc = np.arange(-1.,1.1,0.1)
    Yc = np.arange(0.,2.1,0.1)
    dA = 0.1*0.1
    sigma = 1e-6 #1 microcoulomb/m^2
    dq = sigma*dA
    x = 2
    y = 3
    
    f1 = pointPotential(x,y,dq,Xc,Yc)
    print f1.sum() 

    
    xmin = -1.
    xmax = 1.
    ymin = 0.
    ymax = 2.

    samples_x = xmin + (xmax-xmin)*np.random.random_sample(N);
    samples_y = ymin + (ymax-ymin)*np.random.random_sample(N);
    
    
    approx = pointPotential(x,y,sigma,samples_x,samples_y)

    I = approx.mean()*(xmax-xmin)*(ymax-ymin)
    
    
    #Numerical uncertainty
    Iunc = (ymax-ymin)*(xmax-xmin)*np.sqrt( ((approx**2).mean()-(approx.mean())**2)/N )
    print I,' +/- ',Iunc, "microcoulombs"

    #Plot our function
    plt.xlim(-3.,3.)
    plt.ylim(0.,4.)

    #plot the box
    plt.plot([xmin, xmin], [ymin, ymax], color='k', linestyle='--') 
    plt.plot([xmax, xmax], [ymin, ymax], color='k', linestyle='--') 
    plt.plot([xmin, xmax], [ymax, ymax], color='k', linestyle='--')
    plt.plot(samples_x[1:500],samples_y[1:500],'g.')
    plt.plot(2, 3, 'o', markersize = 5, color = 'r');

    #Plot the average value of the function from the samples on the same graph for comparison:
    plt.plot((xmin,xmax),(approx.mean(),approx.mean()),'r-',label="Average value of function at sample points")
    
    plt.show()
```

```python
sampleMean(100)
sampleMean(1000)
sampleMean(10000)
sampleMean(100000)
sampleMean(1000000)
```
