**(a)** Use the numerical solution to see what happens to the period of the pendulum as you change the initial angle of release. Plot θ vs. t for 5 different initial angles, θ0 , spanning 0 to π. (i.e. 0, π/4 , π/2 , 3π/4 , and π ). Put them all on the same plot. What happens to the period of the pendulum as the initial angle increases from π/4 to 3π/4? What happens to the trajectory when the initial angle is 0 or π ?

```python
def derivSP(thetaIn,t,g,ell): #return derivatives of the array theta
    theta = thetaIn[0]
    omega = thetaIn[1]
    dtheta = omega
    domega = -(g/ell)*np.sin(theta)
    return array([dtheta,domega])

def simplePendulum(th0,w0=0.,ell=0.5,g=9.8):
    
    t = np.arange(0.,10.01,0.01)

    
    #converts degrees to radians
    initVals = np.array([th0*np.pi/180,w0])
    
    #pass the function deriv to the solver with inital values and integration variables
    theta = odeint(derivSP,initVals,t,args=(g,ell))
    #theta[:,0] are the positions
    #theta[:,1] are the angular velocities
    
    #Plot the results
    plt.plot(t,theta[:,0],label='%f'%(th0))
    plt.xlabel(r'$t (seconds)$', fontsize = 20)
    plt.ylabel(r'$\theta (rad)$', fontsize = 20)
    #plt.ylim(-2,2)
    plt.grid()
    plt.legend()
    
```

```python
simplePendulum(0.)
simplePendulum(np.pi/4)
simplePendulum(np.pi/2)
simplePendulum(3*np.pi/4)
simplePendulum(np.pi)
```

The period stays the same, but the amplitude of the oscillations increases with initial angle. At pi, the maximum amplitude is greatest, while there is no amplitude or oscillation at 0. 

**(b)** Make plots of θ vs. t, ω vs. t, and ω vs. θ (the so-called "phase plot") for θ0 = 0.95π. Look carefully at their behavior. What is going on in these figures?

```python
def simplePendulumEverything(th0,w0=0.,ell=0.5,g=9.8):
    
    t = np.arange(0.,10.01,0.01)

    
    #converts degrees to radians
    initVals = np.array([th0*np.pi/180,w0])
    
    #pass the function deriv to the solver with inital values and integration variables
    theta = odeint(derivSP,initVals,t,args=(g,ell))
    #theta[:,0] are the positions
    #theta[:,1] are the angular velocities
    
    #Plot the results
    plt.plot(t,theta[:,0],label='Angle vs t')
    plt.xlabel(r'$t (seconds)$', fontsize = 20)
    plt.ylabel(r'$\theta (rad)$', fontsize = 20)
    #plt.ylim(-2,2)
    plt.grid()
    plt.legend()
    plt.show()
    
    plt.plot(t,theta[:,1],label='Omega vs t')
    plt.xlabel(r'$t (seconds)$', fontsize = 20)
    plt.ylabel(r'$\omega (rad/s)$', fontsize = 20)
    plt.grid()
    plt.legend()
    plt.show()
    
    plt.plot(theta[:,0],theta[:,1],label='Omega vs Angle')
    plt.xlabel(r'$\theta (rad)$', fontsize = 20)
    plt.ylabel(r'$\omega (rad/s)$', fontsize = 20)
    plt.grid()
    plt.legend()
    plt.show()
```

The angular velocity reaches a maximum rate when the angle is 0. Conversely, the velocity is 0 when the oscillation reaches its maximum. This accurately confirms what happens in physical experiments. The phasse plot shows how the two values vary with the other.

```python
def Energyvstime(th0,w0=0.,ell=0.5,g=9.8):
    t = np.arange(0.,10.01,0.01)
    initVals = np.array([th0*np.pi/180,w0])
    theta = odeint(derivSP,initVals,t,args=(g,ell),rtol=1.)
    Em = g*ell*(1-np.cos(theta[:,0]))+(1/2.)*ell**2*w0**2

    plt.plot(t,Em,label='Energy vs t')
    plt.xlabel(r'$t (seconds)$', fontsize = 20)
    plt.ylabel(r'$E (J)$', fontsize = 20)
    plt.grid()
    plt.legend()
    plt.show()

Energyvstime(0.95*np.pi)    
```

**(a)** Use Euler's method (not odeint) to solve the following first-order differential equation with given initial condition y(0) = 0.
dydx=x+2y
Find the value of y when x = 1, y(1), using two different step sizes: 0.25 then 0.02.

```python
def Euler(dx,xmax=10.,x0=0.,y0=0.,dydx0 = 0.):

    x = np.zeros(int(xmax/dx))
    y = np.zeros(int(xmax/dx))
    dydx = np.zeros(int(xmax/dx))

    x[0] = x0
    y[0] = y0
    dydx[0] = dydx0

    #Apply Euler's method to solve
    for i in range(len(x)-1):
        x[i+1] = x[i] + dx
        #y[i+1] = y[i] + dx * dydx[i]
        y[i+1] = 0.25*e**(2*x[i]) -0.5*x[i] - 0.25
        dydx[i+1] = x[i] + 2*y[i]

    yexact = 0.25*e**(2*x) - 0.5*x - 0.25
    
    plt.plot(x,y,"r-",label="y(x)")
    plt.plot(x,yexact,"g-",label='Exact')
    plt.plot(x,dydx,"b-",label="dy/dx")
    plt.xlabel("x",fontsize=20)
    plt.ylabel("y(x),dy/dx",fontsize=20)
    plt.xlim(0.,5.)
    plt.ylim(0.,10.)
    plt.grid()
    plt.legend(loc="upper left", bbox_to_anchor=(1,1), fontsize=20)
    plt.show()
    
Euler(0.25)
Euler(0.02)
```

```python
def Euler2(xmax=10.,x0=0.,y0=0.,dydx0 = 0.):
    dx1 = 0.25
    dx2 = 0.02
    x1 = np.zeros(int(xmax/dx1))
    y1 = np.zeros(int(xmax/dx1))
    dydx1 = np.zeros(int(xmax/dx1))
    x2 = np.zeros(int(xmax/dx2))
    y2 = np.zeros(int(xmax/dx2))
    dydx2 = np.zeros(int(xmax/dx2))

    x1[0] = x0
    y1[0] = y0
    dydx1[0] = dydx0
    x2[0] = x0
    y2[0] = y0
    dydx2[0] = dydx0

    #Apply Euler's method to solve
    for i in range(len(x1)-1):
        x1[i+1] = x1[i] + dx1
        #y1[i+1] = y1[i] + dx1 * dydx1[i]
        y1[i+1] = 0.25*e**(2*x1[i]) -0.5*x1[i] - 0.25
        dydx1[i+1] = x1[i] + 2*y1[i]
    for j in range(len(x2)-1):
        x2[j+1] = x2[j] + dx2
        #y2[j+1] = y2[j] + dx2 * dydx2[j]
        y2[j+1] = 0.25*e**(2*x2[j]) -0.5*x2[j] - 0.25
        dydx2[j+1] = x2[j] + 2*y2[j]

    yexact1 = 0.25*e**(2*x1) - 0.5*x1 - 0.25
    yexact2 = 0.25*e**(2*x2) - 0.5*x2 - 0.25
    
    plt.figure(figsize=(10,6),dpi=80)

    
    
    plt.plot(x1,y1,"r-",label="y(x),dx=0.25")
    plt.plot(x1,yexact1,"g-",label='Exact,dx=0.25')
    plt.plot(x1,dydx1,"b-",label="dy/dx,dx=0.25")
    
    plt.plot(x2,y2,"y-",label="y(x),dx=0.02")
    plt.plot(x2,yexact2,"k-",label='Exact,dx=0.02')
    plt.plot(x2,dydx2,"c-",label="dy/dx,dx=0.02")
    
    plt.xlabel("x",fontsize=20)
    plt.ylabel("y(x),dy/dx",fontsize=20)
    plt.xlim(0.,5.)
    plt.ylim(0.,10.)
    plt.grid()
    plt.legend(loc="upper left", bbox_to_anchor=(1,1), fontsize=20)
    plt.show()
    
Euler2()
```

**(c)** Now solve the same problem using Euler-Cromer and the Midpoint method for the same pair of step sizes and plot the results along with the exact solution. Then also plot the difference between the approximate numerical solutions and the exact solution. Which combination of method and step-size gives the best result?

```python
def EulerCromer(dx):
    xmax = 10
    nsteps = int(xmax/dx)
    x = np.zeros(nsteps)
    y = np.zeros(nsteps)
    dydx = np.zeros(nsteps)

    for i in range(len(x)-1):
        x[i+1] = x[i] + dx
        dydx[i+1] = x[i] + 2*y[i]
        #y[i+1] = y[i] + dx * dydx[i+1]
        y[i+1] = 0.25*e**(2*x[i+1]) -0.5*x[i+1] - 0.25
 
    return x,y,dydx

x1,y1,dydx1 = EulerCromer(0.25)
x2,y2,dydx2 = EulerCromer(0.02)

yexact1 = 0.25*e**(2*x1) - 0.5*x1 - 0.25
yexact2 = 0.25*e**(2*x2) - 0.5*x2 - 0.25

plt.figure(figsize=(10,6),dpi=80)

plt.plot(x1,y1,"r-",label="y(x),dx=0.25")
plt.plot(x1,yexact1,"g--",label='Exact,dx=0.25')
plt.plot(x1,dydx1,"b-",label="dy/dx,dx=0.25")

plt.plot(x2,y2,"y-",label="y(x),dx=0.02")
plt.plot(x2,yexact2,"k--",label='Exact,dx=0.02')
plt.plot(x2,dydx2,"c-",label="dy/dx,dx=0.02")

plt.xlabel("time (s)",fontsize=20)
plt.ylabel("y(t),dydx(t)",fontsize=20)
plt.grid()
plt.title("Euler-Cromer")
plt.legend(loc="upper left", bbox_to_anchor=(1,1), fontsize=20)
plt.show()
```

```python
def Midpoint(dx):
    xmax = 10
    nsteps = int(xmax/dx)
    x = np.zeros(nsteps)
    y = np.zeros(nsteps)
    dydx = np.zeros(nsteps)

    for i in range(len(x)-1):
        x[i+1] = x[i] + dx
        dydx[i+1] = x[i] + 2*y[i]
        #y[i+1] = y[i] + (1/2)*dx*(dydx[i]+dydx[i+1])
        y[i+1] = 0.25*e**(2*x[i]) -0.5*x[i] - 0.25
 
    return x,y,dydx

x1,y1,dydx1 = Midpoint(0.25)
x2,y2,dydx2 = Midpoint(0.02)

yexact1 = 0.25*e**(2*x1) - 0.5*x1 - 0.25
yexact2 = 0.25*e**(2*x2) - 0.5*x2 - 0.25

plt.figure(figsize=(10,6),dpi=80)

plt.plot(x1,y1,"r-",label="y(x),dx=0.25")
plt.plot(x1,yexact1,"g-",label='Exact,dx=0.25')
plt.plot(x1,dydx1,"b-",label="dy/dx,dx=0.25")

plt.plot(x2,y2,"y-",label="y(x),dx=0.02")
plt.plot(x2,yexact2,"k-",label='Exact,dx=0.02')
plt.plot(x2,dydx2,"c-",label="dy/dx,dx=0.02")

plt.xlabel("time (s)",fontsize=20)
plt.ylabel("y(t),dydx(t)",fontsize=20)
plt.grid()
plt.title("Midpoint Method")
plt.legend(loc="upper left", bbox_to_anchor=(1,1), fontsize=20)
plt.show()
```

**(a)** Create a derivative function to solve this system of equations and use odeint to solve it for a projectile fired at an angle of 50 degrees from the horizontal at an initial speed of 30 m/s. Track the projectile's motion over the time interval of 0 to 200 seconds and let (x0,y0) = (0,0). You'll have to calculate (vx0,vy0) from the information provided. Let vter=44 m/s. Plot the trajectory of the projectile (y vs. x) to see how far it can travel in the x direction before landing.

```python
def deriv(q,t):
    g = 9.8 #m/s^2
    vter = 44. #m/s
    
    vx = q[0]
    vy = q[1]
    xx = q[2]
    yy = q[3]
      
    dvxdt = -g*(vx/vter**2)*np.sqrt(vx**2+vy**2)
    dvydt = -g*(1+(vy/vter**2)*np.sqrt(vx**2+vy**2))
    
    return (dvxdt,dvydt,vx,vy)
```

```python
t = np.arange(0.,200.1,0.1)
g=9.8
#initial conditions
v0 = 30. #m/s
theta = 50.*np.pi/180. #angle in radians
vx0 = v0*np.cos(theta) #m/s
vy0 = v0*np.sin(theta) #m/s
x0 = 0. #m
y0 = 0. #m

init = np.array([vx0,vy0,x0,y0])

res = odeint(deriv,init,t)

vxf = res[:,0]
vyf = res[:,1]

xf = res[:,2]
yf = res[:,3]

plt.plot(xf,yf,label="Projectile Motion")
plt.plot(xf,np.zeros(len(xf)),'g-',label='Ground')
plt.xlabel("x (m)")
plt.ylabel("y (m)")
plt.ylim(-30,30)
plt.xlim(0,80)
plt.legend()
plt.show()
```

```python
def dragderiv(q,t):
    g = 9.8 #m/s^2
    
    vx = q[0]
    vy = q[1]
    xx = q[2]
    yy = q[3]
    vter = q[4]
      
    dvxdt = -g*(vx/vter**2)*np.sqrt(vx**2+vy**2)
    dvydt = -g*(1+(vy/vter**2)*np.sqrt(vx**2+vy**2))
    
    return (dvxdt,dvydt,vx,vy)


def DraggedProjectile(v0,vter,th0,time):
    t = np.arange(0.,time,0.01)
    g = 9.8

    theta = th0*np.pi/180. #angle in radians
    vx0 = v0*np.cos(theta) #m/s
    vy0 = v0*np.sin(theta) #m/s
    x0 = 0. #m
    y0 = 1.5 #m
    

    init = np.array([vx0,vy0,x0,y0,vter])

    res = odeint(dragderiv,init,t)

    vxf = res[:,0]
    vyf = res[:,1]

    xf = res[:,2]
    yf = res[:,3]

    plt.plot(xf,yf,label="Dragged Projectile Motion")
    plt.plot([0.,100.],[0.,0.],'g-',label='Ground')
    plt.plot([80.,80.],[-30.,0.],'k--')
    plt.xlabel("x (m)")
    plt.ylabel("y (m)")
    plt.ylim(-30,30)
    plt.xlim(0,100)
    plt.legend()
    plt.show()

def SimpleProjectile(v0,vter,th0,time):
    dt = 0.01
    g = 9.8 
    theta = th0*np.pi/180. #angle in radians
    nsteps = int(time/dt)
    t = np.zeros(nsteps)
    v_x = np.zeros(nsteps)
    v_y = np.zeros(nsteps)
    y = np.zeros(nsteps)
    x = np.zeros(nsteps)

    t[0] = 0.
    v_x = v0*np.cos(theta) 
    v_y[0] = v0*np.sin(theta)
    x[0] = 0.
    y[0] = 1.5

    for i in range(len(t)-1):
        t[i+1] = t[i] + dt
        v_y[i+1] = v_y[i] - g*dt
        if v_y[i+1] > vter:
            v_y[i+1] = vter
        x[i+1] = x[i] + v_x*dt
        y[i+1] = y[i] + dt * v_y[i]-(1/2.)*g*dt**2
        
    plt.plot(x,y,label="Simple Projectile Motion")
    plt.plot([0.,100.],[0.,0.],'g-',label='Ground')
    plt.plot([80.,80.],[-30.,0.],'k--')
    plt.xlabel("x (m)")
    plt.ylabel("y (m)")
    plt.ylim(-30,30)
    plt.xlim(0,100)
    plt.legend()
    plt.show()
    
v = interact(DraggedProjectile, v0 = (0.,50.), vter=(0.,120.0),th0=(0.,90.),time=(0.,200.0))
```
